```python
import geopandas
import shapely
import pandas as pd

df = pd.read_csv("data/clean/parking_garages.csv")
```

```python
from shapely import wkt
gf  = df.copy()
gf.geometry = gf.geometry.apply(wkt.loads)
gf = geopandas.GeoDataFrame(gf)
gf
```

```python
geo = gf.geometry
poly = geo[1]
poly.bounds
#poly
```

```python
b = poly.bounds
minx, miny, maxx, maxy = b
s = shapely.Polygon(shell=((minx, miny), (minx, maxy), (maxx, maxy), (maxx, miny)))
b
```

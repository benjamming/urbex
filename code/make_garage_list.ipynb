```python
to_drop = ['OBJECTID', "CONTACT", "FAC_NUM", "COMPANY", "OPERATOR", "ID", "LDMD", "STR_BLOCK",
            "MEMO", "NOTES", "CITYPARK", "PROPOSED", "PARC","TOTAL_SPCS", "MONTHAVAIL", "PUB_SPC",
            "PHONE_AREA" ]
exclude_strings = ['RATE', 'TAX', 'WEB', "PVT", "SHAPE"]
for string in exclude_strings:
    for col in parking.columns:
        if string in col:
            to_drop.append(col)

redundant = ['ADDRESS_1', "NAME"]
to_drop.extend(redundant)

#to_drop
cleaner = parking.drop(to_drop, axis=1)
cleaner
```

```python
cleaner[cleaner.CLASS != "PUBLIC"]
```

```python
garages = cleaner[cleaner.TYPE=="G"]
garages
```

```python
to_drop_garages = ["TYPE", "CLASS", "TYPE_FAC", "PRIVATE", "PUB_ACCESS", "STREETSORT"]
parking_garages = garages.drop(to_drop_garages, axis=1)
```

```python
pg = parking_garages.reset_index()
pg = pg.rename({'index':'source_index', 'ADDRESS':'address', 'FACILITY':'facility', 'LEVELS':'levels'}, axis=1)
```

```python
pg.to_csv("data/cleaner/parking_garages.csv")
```
